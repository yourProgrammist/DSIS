## MATPLOTLIB

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
fig = plt.figure(figsize=(10, 4))
ax = fig.add_subplot()
x = np.arange(-10, 11, 1) #     x
y = np.array([-a * a for a in x]) #     y
ax.plot(x, y, linewidth=2, marker='+', linestyle='--', color="#000000", label=r'$f(x) = -x^2$')
ax.annotate(r'$top$', xy=(0, 0), xytext=(7.5, 10), arrowprops={'facecolor': 'red', 'linewidth': 1})
y1 = np.array([a * a - 100 for a in x]) 
ax.plot(x, y1, linewidth=2, marker='+', linestyle='-', color="#000000", label=r'$f(x) = x^2 - 100$')
ax.annotate(r'$bottom$', xy=(0, -100), xytext=(7.5, -120), arrowprops={'facecolor': 'red', 'linewidth': 1})
ax.set_xlabel('Ox')
ax.set_ylabel('Oy')
ax.fill_between(-x, y1, color='grey', alpha=0.3)
ax.legend()
ax.grid()
```

```python
from matplotlib.ticker import MultipleLocator, FixedLocator, LogLocator
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot(1, 3, 1)
ax.plot(np.arange(1, 5, 0.2))
ax.set(xlim = (-10, 10), ylim = (-3, 6))
ax.grid()
ax.xaxis.set_major_locator(MultipleLocator(base=5))
ax.yaxis.set_major_locator(MultipleLocator(base=3))

ax2 = fig.add_subplot(1, 3, 2)
ax2.grid()
x = np.arange(-np.pi, np.pi, 0.2)
y = np.sin(x)
plt.plot(x, y)
ax2.xaxis.set_major_locator(FixedLocator([-1, 1, 3]))
ax2.yaxis.set_major_locator(FixedLocator([-1, 0, 1]))

ax3 = fig.add_subplot(1, 3, 3)
ax3.grid()
x = np.arange(-10, 10, 1)
y = np.array([a ** 3 for a in x])
plt.plot(x, y)
ax3.yaxis.set_major_locator(LogLocator(base=25))
```

#  `imshow`

```python
vegetables = ["cucumber", "tomato", "lettuce", "asparagus",
              "potato", "wheat", "barley"]
farmers = ["Farmer Joe", "Upland Bros.", "Smith Gardening",
           "Agrifun", "Organiculture", "BioGoods Ltd.", "Cornylee Corp."]

harvest = np.array([[0.8, 2.4, 2.5, 3.9, 0.0, 4.0, 0.0],
                    [2.4, 0.0, 4.0, 1.0, 2.7, 0.0, 0.0],
                    [1.1, 2.4, 0.8, 4.3, 1.9, 4.4, 0.0],
                    [0.6, 0.0, 0.3, 0.0, 3.1, 0.0, 0.0],
                    [0.7, 1.7, 0.6, 2.6, 2.2, 6.2, 0.0],
                    [1.3, 1.2, 0.0, 0.0, 0.0, 3.2, 5.1],
                    [0.1, 2.0, 0.0, 1.4, 0.0, 1.9, 6.3]])


fig, ax = plt.subplots()
im = ax.imshow(harvest)

ax.set_xticks(np.arange(len(farmers)), labels=farmers)
ax.set_yticks(np.arange(len(vegetables)), labels=vegetables)

plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

for i in range(len(vegetables)):
    for j in range(len(farmers)):
        text = ax.text(j, i, harvest[i, j],
                       ha="center", va="center", color="w")

ax.set_title("Harvest of local farmers (in tons/year)")
fig.tight_layout()
plt.show()
```

# Altair

```python
import seaborn as sns
import altair as alt 

penguins = sns.load_dataset("penguins")
penguins.head()
```

## `mark_bar`

```python
alt.Chart(penguins).mark_bar().encode(
    alt.X('island'),
    y='count()'
).properties(
   height=250, width=450
)
```

## `mark_point`

```python
alt.Chart(penguins).mark_point().encode(
    x='body_mass_g',
    y='flipper_length_mm'
).interactive() #   
```

## `mark_rect`

```python
alt.Chart(penguins).mark_rect().encode(
    x=alt.X('body_mass_g', bin=True),
    y=alt.Y('flipper_length_mm', bin=True),
    color='count()'
)
```

##  `tooltip`

```python
alt.Chart(penguins).mark_point(filled=True).encode(
    alt.X('body_mass_g'),
    alt.Y('flipper_length_mm'),
    alt.OpacityValue(0.7),
    tooltip = [alt.Tooltip('body_mass_g'),
               alt.Tooltip('flipper_length_mm')
              ]
).interactive()
```
